```python
url = 'https://raw.githubusercontent.com/88alexinho/data_analyst_karpov_course/refs/heads/main/2_bookings.csv'
```

1. Импортируйте библиотеку pandas как pd. Загрузите датасет bookings.csv с разделителем ;. Проверьте размер таблицы, типы переменных, а затем выведите первые 7 строк, чтобы посмотреть на данные.

```python
import pandas as pd 

data_booking = pd.read_csv(url, sep=';')
data_booking.head(7)
```

```python
data_booking.info()
```

2. Приведите названия колонок к нижнему регистру и замените пробелы на знак нижнего подчеркивания.


```python
data_booking.rename(columns = lambda x: x.lower().replace(' ', '_'), inplace=True)
data_booking.columns
```

3. Пользователи из каких стран совершили наибольшее число успешных бронирований? Укажите топ-5.

```python
data_booking.groupby(['country'])['is_canceled'].sum(numeric_only=True).sort_values(ascending=False).head(5)
```

4. На сколько ночей в среднем бронируют отели разных типов?


```python
data_booking.groupby(['hotel']).aggregate({'stays_total_nights': 'mean'}) \
            .sort_values(by='stays_total_nights', ascending=False)
```

5. Иногда тип номера, полученного клиентом (assigned_room_type), отличается от изначально забронированного (reserved_room_type). Такое может произойти, например, по причине овербукинга. Сколько подобных наблюдений встретилось в датасете?

```python
data_booking[data_booking['assigned_room_type'] != data_booking['reserved_room_type']].shape[0]
```

6. Проанализируйте даты запланированного прибытия. 

* На какой месяц чаще всего успешно оформляли бронь в 2016? 
* Изменился ли самый популярный месяц в 2017?


```python
data_booking.groupby('arrival_date_year').arrival_date_month.max()
```

* Сгруппируйте данные по годам и проверьте, на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в каждый из периодов

```python
city_hotel_data = data_booking[data_booking['hotel'] == 'City Hotel']

cancelled_group = city_hotel_data.groupby(['arrival_date_year', 'arrival_date_month', 'is_canceled']) \
    .agg({'hotel': 'count'}) \
    .reset_index()

cancelled_data = cancelled_group[cancelled_group['is_canceled'] == 1]

top_cancel_months = cancelled_data.groupby('arrival_date_year') \
    .apply(lambda x: x.nlargest(1, 'hotel')) \
    .reset_index(drop=True)

print(top_cancel_months)
```

7. Посмотрите на числовые характеристики трёх переменных: adults, children и babies. Какая из них имеет наибольшее среднее значение?

```python
data_booking[['adults', 'children', 'babies']].describe()
```

У adults наибольшее среднее значение

8. Создайте колонку total_kids, объединив children и babies. Для отелей какого типа среднее значение переменной оказалось наибольшим?

```python
data_booking['total_kids'] = data_booking['children'] + data_booking['babies']

data_booking.groupby('hotel').aggregate({'total_kids': 'mean'})
```

Для типа Resort Hotel

9. Создайте переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), в противном случае – False. 

Посчитайте отношение количества ушедших пользователей к общему количеству клиентов, выраженное в процентах (churn rate). Укажите, среди какой группы показатель выше.

```python
data_booking['has_kids'] = data_booking.apply(lambda x: x['total_kids'] > 0, axis=1)
data_booking
```

```python
data_booking['has_kids'].value_counts()
```

```python
data_booking[['has_kids', 'is_canceled']].value_counts() / data_booking['has_kids'].value_counts() * 100
```

Churn rate - 37.2%
